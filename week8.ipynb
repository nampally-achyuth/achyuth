```
class person:
    def __init__(self,name):
        self.name=name
    def display(self):
        print(self.name)
class student(person):
    def __init__(self,name,rollno):
      super().__init__(name)
      self.rollno=rollno
    def show(self):
        print(self.rollno)
s1=student("Achyuth",60)
s1.display()
s1.show()
```
