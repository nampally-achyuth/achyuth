```
class person:
    def __init__(self,name):
        self.name=name
    def display(self):
        print(self.name)
class student(person):
    def __init__(self,name,rollno):
      super().__init__(name)
      self.rollno=rollno
    def show(self):
        print(self.rollno)
s1=student("Achyuth",60)
s1.display()
s1.show()
```

```
class animal():
    def sound(self):
        print("animal sound")
class dog(animal):
    def sound(self):
        print("dog barks")
class cat(animal):
    def sound(self):
        print("cat meows")
class duck(animal):
    def sound(self):
        print("duck quacks")
d1=dog()
d1.sound()
c1=cat()
c1.sound()
d2=duck()
d2.sound()
```
